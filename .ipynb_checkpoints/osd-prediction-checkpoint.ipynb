```python
import json,os
```

```python
import pandas as pd
```

```python
import datetime as dt
```

```python
## Load dataset
DATA_DIR = "dataset/ssd/"    
datasets = os.listdir(DATA_DIR)
```

```python
## Create dict of dataset 
datasets_dict = []
for i in datasets: 
    data_dict = {}
    data_date = i.split('-')
    data_dict["date"] = data_date[2]+"-"+data_date[3]+"-"+data_date[4] 
    f = open(DATA_DIR+i) 
    osd_data = json.load(f)
    for k in osd_data["nodes"]: 
        if 'type' in k.keys() and k["type"] == "osd" and k["reweight"] != 0 and k["status"] == "up":
            data_dict[f"osd_{k['id']}"] = k['utilization']
    datasets_dict.append(data_dict)
```

```python
## Convert it into dataframe 
df = pd.DataFrame.from_dict(datasets_dict)
```

```python
## Set date col with new format 
df['date'] = pd.to_datetime(df['date'], format='%d-%m-%Y')
```

```python
df.set_index('date',inplace=True)
```

```python
df.sort_values("date",inplace=True)
```

```python
df
```

```python
## Remove the old osds with NaN value 
for i in df.keys(): 
    if df[i].tail(7).isnull().sum() == 7:
        df.drop(columns=[i],inplace=True)
    else:
        #marked as new osd
        if df[i].head(7).isnull().sum() == 7:
            df[i].fillna(0, inplace=True) 
            
        df[i].fillna(df[i].mean(), inplace=True)
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
pre = LinearRegression() 
dateTomorrow = df.tail(1).index[0]
```

```python
if "date_toordinal" not in df.keys():
    date_toordinal = []
    for i in df.index:
        date_toordinal.append(i.toordinal()) 
    df.insert(0, "date_toordinal", date_toordinal, True)    
```

```python
x = np.array(df["date_toordinal"]).reshape((-1, 1))
```

```python
df.tail(10)
```

```python
poor_score = {}
```

```python
for i in df.keys(): 
    if df[i].tail(7).sum() != 0:
        if df[i].isnull().sum():
            print(f"{i} have null data for 1 week ,skip create prediction") 
            continue
        
        #if (df[i].tail(10).sum() - df[i].tail(10).median()) / df[i].tail(10).sum() < 0.80:
        #    print(f"{i} have inconsistent data for 1 week ,skip create prediction") 
        #    continue
            
        y = np.array(df[i]) 
        pre.fit(x,y)  
        predict_x = pre.predict([[dateTomorrow.toordinal()]])  
        score_x = pre.score(x,y) 
        if score_x < 0.5:
            poor_score[i] = df[i].copy()
        print(f"{i} have {predict_x} on {dateTomorrow} with score {score_x}")  
```

```python
df_poor_score = pd.DataFrame.from_dict(poor_score)
```

```python
if len(poor_score) > 0: 
    import matplotlib.pyplot as plt

    x1 = df_poor_score.index
    for i in df_poor_score.keys():
        plt.plot(x1, df_poor_score[i], label = df_poor_score[i].name)

    plt.xlabel('Date')
    plt.ylabel('Value')
    plt.title('Osd with low score')
    plt.savefig('poor_score.png')
    plt.legend()
    plt.show()
```
